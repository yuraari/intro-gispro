### Geospatial Example: Haversine Function

```python
from math import radians, sin, cos, sqrt, atan2
```

```python
def haversine(lat1, lon1, lat2, lon2):
    R = 6371.0  # Earth radius in kilometers
    dlat = radians(lat2 - lat1)
    dlon = radians(lon2 - lon1)
    a = (
        sin(dlat / 2) ** 2
        + cos(radians(lat1)) * cos(radians(lat2)) * sin(dlon / 2) ** 2
    )
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    distance = R * c
    return distance


# Example usage
distance = haversine(35.6895, 139.6917, 34.0522, -118.2437)
print(f"Distance: {distance:.2f} km")
```

### Function with Default Values and Geospatial Application

```python
def haversine(lat1, lon1, lat2, lon2, radius=6371.0):
    dlat = radians(lat2 - lat1)
    dlon = radians(lon2 - lon1)
    a = (
        sin(dlat / 2) ** 2
        + cos(radians(lat1)) * cos(radians(lat2)) * sin(dlon / 2) ** 2
    )
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    distance = radius * c
    return distance


# Example usage in kilometers
distance_km = haversine(35.6895, 139.6917, 34.0522, -118.2437)
print(f"Distance in kilometers: {distance_km:.2f} km")

# Example usage in miles (radius of Earth is approximately 3958.8 miles)
distance_miles = haversine(35.6895, 139.6917, 34.0522, -118.2437, radius=3958.8)
print(f"Distance in miles: {distance_miles:.2f} miles")
```

### Processing Multiple Coordinates

```python
def batch_haversine(coord_list):
    distances = []
    for i in range(len(coord_list) - 1):
        lat1, lon1 = coord_list[i]
        lat2, lon2 = coord_list[i + 1]
        distance = haversine(lat1, lon1, lat2, lon2)
        distances.append(distance)
    return distances


# Example usage
coordinates = [(35.6895, 139.6917), (34.0522, -118.2437), (40.7128, -74.0060)]
distances = batch_haversine(coordinates)
print(f"Distances: {distances}")
```

### Advanced Parameter Handling

#### Variable Arguments with \*args

```python
def average(*numbers):
    return sum(numbers) / len(numbers)


# Example usage
print(average(10, 20, 30))  # 20.0
print(average(5, 15, 25, 35))  # 20.0
```

#### Keyword Arguments with \*\*kwargs

```python
def describe_point(latitude, longitude, **kwargs):
    description = f"Point at ({latitude}, {longitude})"

    # Add optional keyword arguments to the description
    for key, value in kwargs.items():
        description += f", {key}: {value}"

    return description


# Example usage
print(describe_point(35.6895, 139.6917, name="Tokyo", population=37400000))
print(describe_point(34.0522, -118.2437, name="Los Angeles", state="California"))
```

## Classes: Organizing Data and Behavior Together

### Defining a Simple Class

```python
class Point:
    def __init__(self, latitude, longitude, name=None):
        self.latitude = latitude
        self.longitude = longitude
        self.name = name

    def __str__(self):
        return f"{self.name or 'Point'} ({self.latitude}, {self.longitude})"


# Example usage
point1 = Point(35.6895, 139.6917, "Tokyo")
print(point1)
```

### Adding Methods to a Class

```python
class Point:
    def __init__(self, latitude, longitude, name=None):
        self.latitude = latitude
        self.longitude = longitude
        self.name = name

    def distance_to(self, other_point):
        return haversine(
            self.latitude, self.longitude, other_point.latitude, other_point.longitude
        )


# Example usage
point1 = Point(35.6895, 139.6917, "Tokyo")
point2 = Point(34.0522, -118.2437, "Los Angeles")
print(
    f"Distance from {point1.name} to {point2.name}: {point1.distance_to(point2):.2f} km"
)
```

## Combining Functions and Classes

```python
class Route:
    def __init__(self, points):
        self.points = points

    def total_distance(self):
        total_dist = 0
        for i in range(len(self.points) - 1):
            total_dist += self.points[i].distance_to(self.points[i + 1])
        return total_dist


# Example usage
route = Route([point1, point2])
print(f"Total distance: {route.total_distance():.2f} km")
```

### Exercise 2: Batch Distance Calculation

```python
# Create a function batch_distance_calculation that accepts a list of coordinate pairs and returns a
# list of distances between consecutive pairs.
# • Test the function with a list of coordinates representing several cities.
# • Add an optional parameter to specify the units (kilometers or miles) with a default value.

def batch_distance_calculation(coord_list, units="kilometers"):
    """Calculates distances between consecutive coordinate pairs in a list, with optional units."""
    distances = []
    # Determine the radius based on units
    if units.lower() == "miles":
        radius = 3958.8  # Earth radius in miles
    elif units.lower() == "kilometers":
        radius = 6371.0  # Earth radius in kilometers
    else:
        raise ValueError("Units must be 'kilometers' or 'miles'")

    for i in range(len(coord_list) - 1):
        lat1, lon1 = coord_list[i]
        lat2, lon2 = coord_list[i + 1]
        distance = haversine(lat1, lon1, lat2, lon2, radius=radius)
        distances.append(distance)
    return distances


# Test the function with a list of coordinates representing several cities.
cities_coords = [
    (51.5074, 0.1278),  # London
    (48.8566, 2.3522),  # Paris
    (40.7128, -74.0060), # New York
    (35.6895, 139.6917) # Tokyo
]

print("\nDistances between consecutive cities (kilometers):")
k_distances = batch_distance_calculation(cities_coords, units="kilometers")
for i, dist in enumerate(k_distances):
    print(f"  Distance {i+1}: {dist:.2f} km")

print("\nDistances between consecutive cities (miles):")
m_distances = batch_distance_calculation(cities_coords, units="miles")
for i, dist in enumerate(m_distances):
    print(f"  Distance {i+1}: {dist:.2f} miles")
```

```python

# Define a Point class to represent a geographic point with attributes latitude , longitude , and
# name .
# • Add a method distance_to that calculates the distance from one point to another.
# • Include a __str__ method that provides a readable representation of the point.
# • Instantiate several Point objects and calculate the distance between them.
# • Create a method is_in_northern_hemisphere that returns True if the point is north of the equator.Define a Point class to represent a geographic point with attributes latitude , longitude , and
# name .
# • Add a method distance_to that calculates the distance from one point to another.
# • Include a __str__ method that provides a readable representation of the point.
# • Instantiate several Point objects and calculate the distance between them.
# • Create a method is_in_northern_hemisphere that returns True if the point is north of the equator.
```

```python
class Point:
    def __init__(self, latitude, longitude, name="Unnamed Point"):
        self.latitude = latitude
        self.longitude = longitude
        self.name = name

    def distance_to(self, other_point):
        """Calculates the distance between this point and another point using the haversine formula."""
        # The haversine function is already defined in a previous cell (cell 10)
        return haversine(self.latitude, self.longitude, other_point.latitude, other_point.longitude)

    def __str__(self):
        """Provides a readable string representation of the point."""
        return f"{self.name} ({self.latitude:.4f}, {self.longitude:.4f})"

    def is_in_northern_hemisphere(self):
        """Returns True if the point is north of the equator (latitude > 0)."""
        return self.latitude > 0


# Instantiate several Point objects
point_london = Point(51.5074, 0.1278, "London")
point_paris = Point(48.8566, 2.3522, "Paris")
point_new_york = Point(40.7128, -74.0060, "New York")
point_buenos_aires = Point(-34.6037, -58.3816, "Buenos Aires")

print(f"Point 1: {point_london}")
print(f"Point 2: {point_paris}")
print(f"Point 3: {point_new_york}")
print(f"Point 4: {point_buenos_aires}")

# Calculate distances between them
distance_london_paris = point_london.distance_to(point_paris)
print(f"\nDistance between {point_london.name} and {point_paris.name}: {distance_london_paris:.2f} km")

distance_new_york_london = point_new_york.distance_to(point_london)
print(f"Distance between {point_new_york.name} and {point_london.name}: {distance_new_york_london:.2f} km")

# Check hemispheres
print(f"\n{point_london.name} in Northern Hemisphere: {point_london.is_in_northern_hemisphere()}")
print(f"{point_buenos_aires.name} in Northern Hemisphere: {point_buenos_aires.is_in_northern_hemisphere()}")
```
